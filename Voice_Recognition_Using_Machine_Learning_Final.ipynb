## Using ANN Model

```python
from sklearn import preprocessing

# label encoding
le = preprocessing.LabelEncoder()

y_train_le = le.fit_transform(y_train)
y_test_le = le.fit_transform(y_test)
```

```python
# Importing required layers
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv1D, MaxPooling1D
```

```python
ann_classifier = Sequential()
# first hidden layer and input layer
ann_classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 20))

# second hidden layer
ann_classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# output layer
ann_classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# compiling the optimizer and loss function according to the data
ann_classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# fitting the ANN
ann_classifier.fit(X_train, y_train_le, batch_size = 10, epochs = 200,verbose = 0)
```

### Metrics

```python
score, acc = ann_classifier.evaluate(X_train, y_train_le,
                            batch_size=10)
print('Train score:', score)
print('Train accuracy:', acc)


# predicting the test results
y_pred = ann_classifier.predict(X_test)
y_pred = (y_pred > 0.5)


score, acc = ann_classifier.evaluate(X_test, y_test_le,
                            batch_size=10)
print('Test score:', score)
print('Test accuracy:', acc)

# confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test_le, y_pred)
print(cm)
```

```python
# f1_score
from sklearn.metrics import f1_score
# f1 score metrics
f1_score_value = f1_score(y_test_le,y_pred,average='weighted')
print("F1 score : ", f1_score_value)
```
