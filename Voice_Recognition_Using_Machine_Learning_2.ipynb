# Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(data.iloc[ : , : -1].values, data.iloc[ : ,  -1].values, test_size=0.2)
```

```python
X_train
```

```python
from sklearn.svm import SVC
classifier1 = SVC(kernel = 'rbf')
classifier1.fit(X_train, y_train)
print('Support Vector Machine:')

print('Accuracy of SVM: {:.2f}'.format(classifier1.score(X_test,y_test)))
```

# Machine Learning Models

## Using Logistic regression

```python
# importing logistic regression from scikit learn
from sklearn.linear_model import LogisticRegression
```

```python
# logistic regression model
log_model = LogisticRegression()
```

```python
# fitting the model
log_model.fit(X_train,y_train)
```

```python
# prediction on test data
y_predictions = log_model.predict(X_test)
```

### Metrics

```python
# importing metrics for evaulation of model
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
# classification report of the model
print(classification_report(y_test,y_predictions))
```

```python
# confusion matrix
print("Confusion matrix : \n", confusion_matrix(y_test,y_predictions))
```

```python
# f1_score
from sklearn.metrics import f1_score
# f1 score metrics
f1_score_value = f1_score(y_test,y_predictions,average='weighted')
print("F1 score : ", f1_score_value)
```

## Using Random Forest 

```python
# importing random forest
from sklearn.ensemble import RandomForestClassifier
```

```python
# random forest classifier
randomForestModel = RandomForestClassifier()
```

```python
# fitting the data into the model
randomForestModel.fit(X_train,y_train)
```

```python
# predicting the data
predictions = randomForestModel.predict(X_test)
```

### Metrics

```python
# classification report of the model
print(classification_report(y_test,predictions))
```

```python
print("Confusion matrix : \n",confusion_matrix(y_test,predictions))
```

```python
# f1 score metrics of model
f1_score_value = f1_score(y_test,predictions,average='weighted')
print("F1 score : ", f1_score_value)
```
