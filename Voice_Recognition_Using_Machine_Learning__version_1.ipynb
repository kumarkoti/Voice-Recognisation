```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

%matplotlib inline
```

```python


data = pd.read_csv('/content/drive/MyDrive/content/voice.csv')
data.head()

```

```python
data.corr()
```

```python
print("Total numbers of labels in the dataste : {}".format(data.shape[0]))
print("Count of male category: {}".format(data[data.label == 'male'].shape[0]))
print("Count of female category: {}".format(data[data.label == 'female'].shape[0]))
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(data.iloc[ : , : -1].values, data.iloc[ : ,  -1].values, test_size=0.2)
```

```python
X_train
```

```python
from sklearn.svm import SVC
classifier1 = SVC(kernel = 'rbf')
classifier1.fit(X_train, y_train)
print('Support Vector Machine:')

print('Accuracy of SVM: {:.2f}'.format(classifier1.score(X_test,y_test)))
```

# Machine Learning Models

## Using Logistic regression

```python
# importing logistic regression from scikit learn
from sklearn.linear_model import LogisticRegression
```

```python
# logistic regression model
log_model = LogisticRegression()
```

```python
# fitting the model
log_model.fit(X_train,y_train)
```

```python
# prediction on test data
y_predictions = log_model.predict(X_test)
```

### Metrics

```python
# importing metrics for evaulation of model
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
# classification report of the model
print(classification_report(y_test,y_predictions))
```

```python
# confusion matrix
print("Confusion matrix : \n", confusion_matrix(y_test,y_predictions))
```

```python
# f1_score
from sklearn.metrics import f1_score
# f1 score metrics
f1_score_value = f1_score(y_test,y_predictions,average='weighted')
print("F1 score : ", f1_score_value)
```

## Using K Nearest neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# K nearest neighbour model
knn_model = KNeighborsClassifier(n_neighbors=2)

# fitting the traning data
knn_model.fit(X_train,y_train)

# predicting the test data
y_pred = knn_model.predict(X_test)
```

```python
# accuracy score
accuracy_score_value = accuracy_score(y_test,y_pred)
print("Accuracy Score : ", accuracy_score_value)
```

### Metrics

```python
# metrics of the model
print("Confusion matrix : \n",confusion_matrix(y_test,y_pred))
```

```python
# classification report of model
print(classification_report(y_test,y_pred))
```

```python
# f1 score metrics of model
f1_score_value = f1_score(y_test,y_pred,average='weighted')
print("F1 score : ", f1_score_value)
```
